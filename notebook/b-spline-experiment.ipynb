# B-splines and other experiments for CS445 final project

```python
import numpy as np
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([ 0. ,  1.2,  1.9,  -3.2,  4. ,  6.5])
y = np.array([ 0. ,  2.3,  -3. ,  -4.3,  2.9,  3.1])
t_orig = np.array([ 0. , 1.  ,  2. ,  3. ,  4. ,  5.])

N = 100
tmin, tmax = t_orig.min(), t_orig.max()
tt = np.linspace(tmin, tmax, N)


#-----------------------------------------------------
# x coords
#-----------------------------------------------------
tx, cx, kx = interpolate.splrep(t_orig, x, s=0, k=4)
print('''\
tx: {}
cx: {}
kx: {}
'''.format(tx, cx, kx))
spline_x = interpolate.BSpline(tx, cx, kx, extrapolate=False)

plt.plot(t_orig, x, 'bo', label='Original points')
plt.plot(tt, spline_x(tt), 'r', label='BSpline')
plt.grid()
plt.legend(loc='best')
plt.show()

#-----------------------------------------------------
# y coords
#-----------------------------------------------------
ty, cy, ky = interpolate.splrep(t_orig, y, s=0, k=4)
print('''\
ty: {}
cy: {}
ky: {}
'''.format(ty, cy, ky))
spline_y = interpolate.BSpline(ty, cy, ky, extrapolate=False)

plt.plot(t_orig, y, 'bo', label='Original points')
plt.plot(tt, spline_y(tt), 'r', label='BSpline')
plt.grid()
plt.legend(loc='best')
plt.show()

#-----------------------------------------------------
# combined x,y coords
#-----------------------------------------------------
plt.plot(x, y, 'bo', label='Original points')
plt.plot(spline_x(tt), spline_y(tt), 'r', label='BSpline')
plt.grid()
plt.legend(loc='best')
plt.show()

#-----------------------------------------------------
# x,y coords standard plot linear interpolation
#-----------------------------------------------------
plt.plot(x, y, 'bo', label='Original points')
plt.plot(x, y, 'r', label='Standard linear interpolation')
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
x_coords = []
y_coords = []
for i in range(100):
    x = i / 100
    y = 1.0 - abs(math.cos(100 * x / (40 * (1.0 - 0.9*(x))))) * ( (1 - x) ** 1.2)
    x_coords.append(float(x))
    y_coords.append(y)


plt.plot(x_coords, y_coords, 'bo', label='Original alpha')
plt.plot(x_coords, y_coords, 'r', label='New alpha prime value with bounce effect')
plt.grid()
plt.legend(loc='best')
plt.show()
```
